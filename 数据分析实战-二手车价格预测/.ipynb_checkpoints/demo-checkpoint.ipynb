```python
data['accidentCount'] = data['accidentCount'].fillna(0)
data.isnull().sum()
```

```python
data.dropna(how='any',inplace=True)
data.isnull().sum()
```

```python
data.info()
```

```python
data.head()
```

```python
data['accidentCount'] = data['accidentCount'].astype(int)
```

```python
data = pd.get_dummies(data, columns=["bodyTypeName"], drop_first=True)
# data = pd.get_dummies(data, columns=["exteriorColorName"], drop_first=True)
# data = pd.get_dummies(data, columns=["brand"], drop_first=True)
data.head()
```

```python
data['exteriorColorName'] = data['exteriorColorName'].replace(['BLACK','BLUE','GRAY','RED','SILVER','UNKNOWN','WHITE'],[0,1,2,3,4,5,6])
```

```python
data.drop(['listingTitle'],axis=1,inplace=True)
```

```python
data['localizedTransmission'] = data['localizedTransmission'].replace(to_replace=r'.*?Automatic', value='Automatic', regex=True)
data['localizedTransmission'] = data['localizedTransmission'].replace(to_replace=r'.*?Manual', value='Manual', regex=True)
data['localizedTransmission'] = data['localizedTransmission'].replace(to_replace=r'7-Speed Dual Clutch', value='Automatic', regex=True)
data['localizedTransmission'] = data['localizedTransmission'].replace(to_replace=r'Continuously Variable Transmission', value='Automatic', regex=True)
data['localizedTransmission'] = data['localizedTransmission'].replace(['Automatic','Manual'],[0,1])
data['localizedTransmission'].value_counts()
```

```python
def data_fractionation(x):
    x =str(x)
    x = x.split(",")
    x = x[-1]
    x = x.split(" ")
    x = x[0]
    return x

data['localizedFuelEconomy'] = data['localizedFuelEconomy'].apply(data_fractionation)
data['localizedFuelEconomy'] = data['localizedFuelEconomy'].astype(int)
```

```python
#油耗和价格之间的关系
sns.regplot(x='localizedFuelEconomy',y='price', data=data)
plt.show()
```

```python
data4 = data.drop(['brand','exteriorColorName'], axis=1)
plt.subplots(figsize = (20,9))
sns.heatmap(data4.corr(), annot=True,cmap="PiYG")
```

```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
```

```python
data1 = data.drop(['price'],axis=1)
target = data['price']
train_x,test_x,train_y,test_y = train_test_split(data1,target,test_size=0.25)
```

```python
ss = preprocessing.StandardScaler()
train_ss_x = ss.fit_transform(train_x)
test_ss_x = ss.transform(test_x)
```

```python
RFR = RandomForestRegressor()
RFR.fit(train_ss_x,train_y)
score = cross_val_score(RFR, train_ss_x, train_y, cv=10 )
score
```

```python
data2 = pd.DataFrame(RFR.feature_importances_, index=data.drop('price', axis=1).columns, columns=['importance']).sort_values(by='importance')
data2
```

```python
data2
sns.barplot(x=data2.index,y='importance', data=data2)
plt.xticks(rotation=60)
plt.show()
```

```python
y_pred = RFR.predict(test_ss_x)
y_pred
```

```python
test_x['predict_price'] = y_pred
test_x['old_price'] = test_y
test_x
```

```python
import matplotlib.pyplot as plt
data3 =  test_x.reset_index(drop=True) 
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.figure(figsize=(20, 12))
plt.plot(data3['old_price'], label='old_price', color='g')
plt.plot(data3['predict_price'],label='predict_price',   color='r')
plt.grid(True)
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans
data2 = data.drop(['price','carYear'],axis=1)
#采用的是传统的K-means算法，“auto’会根据数据的特点自动选择是”full“还是”elkan“
kmeans = KMeans(n_clusters=3,init='k-means++',n_init=10,max_iter=300,tol=0.0001,algorithm='auto')
#规范化到[0,1]空间
min_max_scaler = preprocessing.StandardScaler()
train_x = min_max_scaler.fit_transform(data2)
#kmean算法
kmeans.fit(train_x)
predict_y = kmeans.predict(train_x)
data['price category'] = predict_y
data2.head(20)
```
