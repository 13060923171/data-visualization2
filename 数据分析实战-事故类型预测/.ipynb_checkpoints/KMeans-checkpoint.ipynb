```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
n_data = 400
n_cluster = 4
# generate training data
X, y = make_blobs(n_samples=n_data, centers=n_cluster, cluster_std=0.60, random_state=0)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap='rainbow')
plt.show()
```

```python
centers = X[np.random.randint(0,400,4)]
centers
```

```python
labels = np.ones(400,dtype = 'int')
labels.shape
```

```python
for b in range(400):
    labels[b] = np.argmin(np.sum(np.abs(X[b] - centers) ** 2,axis = 1))

plt.scatter(X[:, 0], X[:, 1], c=labels, s=40, cmap='rainbow')
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=60)
plt.show()
```

```python
for c in range(4):
    centers[c]= np.mean(X[labels == c], axis = 0)
    
plt.scatter(X[:, 0], X[:, 1], c=labels, s=40, cmap='rainbow')
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=60)
plt.show()
```

```python
X.shape[0]
```

```python
def KM(X, k =4, Max_Iter = 10):
    n = X.shape[0]
    centers = X[np.random.randint(0,n,k)]
    labels = np.ones(n,dtype = 'int')
    for a in range(Max_Iter):
        for b in range(n):
            labels[b] = np.argmin(np.sum(np.abs(X[b] - centers) ** 2,axis = 1))
        for c in range(k):
            centers[c]= np.mean(X[labels == c], axis = 0)
    plt.scatter(X[:, 0], X[:, 1], c=labels, s=40, cmap='rainbow')
    plt.scatter(centers[:, 0], centers[:, 1], c='black', s=60)
    plt.show()
    
```
