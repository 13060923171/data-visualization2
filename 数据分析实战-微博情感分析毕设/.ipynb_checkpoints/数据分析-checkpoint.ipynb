```python
#ÂÆö‰πâÊú∫Ê¢∞ÂéãÁº©ÂáΩÊï∞
def yasuo(st):
    for i in range(1,int(len(st)/2)+1):
        for j in range(len(st)):
            if st[j:j+i] == st[j+i:j+2*i]:
                k = j + i
                while st[k:k+i] == st[k+i:k+2*i] and k<len(st):   
                    k = k + i
                st = st[:j] + st[k:]    
    return st
```

```python
comment_text = comment_text.apply(yasuo)
comment_text
```

```python
#Êï∞ÊçÆÂ§ÑÁêÜÂ∫ì
import jieba 

#ÂèØËßÜÂåñÂ∫ì
import stylecloud
from IPython.display import Image
# ÂÆö‰πâÂàÜËØçÂáΩÊï∞
def get_cut_words(content_series):
    # ËØªÂÖ•ÂÅúÁî®ËØçË°®
    stop_words = [] 
    
    with open("stopwords_cn.txt", 'r', encoding='utf-8') as f:
        lines = f.readlines()
        for line in lines:
            stop_words.append(line.strip())            

    # ÂàÜËØç
    word_num = jieba.lcut(content_series.str.cat(sep='„ÄÇ'), cut_all=False)

    # Êù°‰ª∂Á≠õÈÄâ
    word_num_selected = [i for i in word_num if i not in stop_words and len(i)>=2]
    return word_num_selected
```

```python
# coding=utf-8
import jieba
import re
import time
from collections import Counter

#------------------------------------‰∏≠ÊñáÂàÜËØç------------------------------------
cut_words = ""
all_words = ""
f = open('fenci.txt', 'w', encoding='utf-8')

for line in comment_text:
    line = line.strip('\n')
    #ÂÅúÁî®ËØçËøáÊª§
    line = re.sub('[0-9‚Äô!"#$%&\'()*+,-./:;<=>?@Ôºå„ÄÇ?‚òÖ„ÄÅ‚Ä¶„Äê„Äë„Ää„ÄãÔºü‚Äú‚Äù‚Äò‚ÄôÔºÅ[\\]^_`{|}~\s]+', "", line)
    seg_list = jieba.cut(line, cut_all=False)
    cut_words = (" ".join(seg_list))

    #ËÆ°ÁÆóÂÖ≥ÈîÆËØç
    all_words = cut_words.split()
    c = Counter()
    for x in all_words:
        if len(x)>1 and x != '\r\n':
            c[x] += 1
    output = ""
    for (k,v) in c.most_common():
        #print("%s:%d"%(k,v))
        output += k + " "
    
    f.write(output+"\n")
else:
    f.close()
```

```python
#ËÅöÁ±ª
# coding=utf-8
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer



#Á¨¨‰∏ÄÊ≠• ËÆ°ÁÆóTFIDF
# ÊñáÊ°£È¢ÑÊñô Á©∫Ê†ºËøûÊé•
corpus = []

# ËØªÂèñÈ¢ÑÊñô ‰∏ÄË°åÈ¢ÑÊñô‰∏∫‰∏Ä‰∏™ÊñáÊ°£
for line in open('fenci.txt', 'r',encoding='utf-8').readlines():
    corpus.append(line.strip())
# Â∞ÜÊñáÊú¨‰∏≠ÁöÑËØçËØ≠ËΩ¨Êç¢‰∏∫ËØçÈ¢ëÁü©Èòµ Áü©ÈòµÂÖÉÁ¥†a[i][j] Ë°®Á§∫jËØçÂú®iÁ±ªÊñáÊú¨‰∏ãÁöÑËØçÈ¢ë
vectorizer = CountVectorizer()

# ËØ•Á±ª‰ºöÁªüËÆ°ÊØè‰∏™ËØçËØ≠ÁöÑtf-idfÊùÉÂÄº
transformer = TfidfTransformer()

# Á¨¨‰∏Ä‰∏™fit_transformÊòØËÆ°ÁÆótf-idf Á¨¨‰∫å‰∏™fit_transformÊòØÂ∞ÜÊñáÊú¨ËΩ¨‰∏∫ËØçÈ¢ëÁü©Èòµ
tfidf = transformer.fit_transform(vectorizer.fit_transform(corpus))
# Ëé∑ÂèñËØçË¢ãÊ®°Âûã‰∏≠ÁöÑÊâÄÊúâËØçËØ≠
word = vectorizer.get_feature_names()

# Â∞Ütf-idfÁü©ÈòµÊäΩÂèñÂá∫Êù• ÂÖÉÁ¥†w[i][j]Ë°®Á§∫jËØçÂú®iÁ±ªÊñáÊú¨‰∏≠ÁöÑtf-idfÊùÉÈáç
weight = tfidf.toarray()

# ÊâìÂç∞ÁâπÂæÅÂêëÈáèÊñáÊú¨ÂÜÖÂÆπ
print('Features length: ' + str(len(word)))

#Á¨¨‰∫åÊ≠• ËÅöÁ±ªKmeans

print('Start Kmeans:')
from sklearn.cluster import KMeans

clf = KMeans(n_clusters=3)
print(clf)
pre = clf.fit_predict(weight)
print(pre)

# ‰∏≠ÂøÉÁÇπ
print(clf.cluster_centers_)
print(clf.inertia_)

#Á¨¨‰∏âÊ≠• ÂõæÂΩ¢ËæìÂá∫ ÈôçÁª¥

from sklearn.decomposition import PCA

pca = PCA(n_components=3)  # ËæìÂá∫‰∏§Áª¥
newData = pca.fit_transform(weight)  # ËΩΩÂÖ•NÁª¥

x = [n[0] for n in newData]
y = [n[1] for n in newData]
plt.figure(figsize=(9,6),dpi = 300)
plt.rcParams['font.sans-serif'] = ['SimHei']  # ÊîØÊåÅ‰∏≠Êñá
plt.rcParams['axes.unicode_minus'] = False
plt.scatter(x, y, c=pre, s=100)
plt.title("ËØçÊÄßËÅöÁ±ªÂõæ")
plt.savefig('ËØçÊÄßËÅöÁ±ªÂõæ.jpg')
plt.show()
```

```python
from snownlp import SnowNLP
from snownlp import sentiment
from snownlp.sentiment import Sentiment
from snownlp import sentiment
from tqdm import tqdm
import re
# ÊåâÊñáÁ´†ÁöÑËØ¥Ê≥ïÔºåÂàùÊ≠•Á≠õÈÄâËØ≠ÊñôÔºåÂ§ß‰∫é0.8ÁöÑÂΩíÂÖ•ÁßØÊûÅÔºåÂ∞è‰∫é0.3ÁöÑÂΩíÂÖ•Ê∂àÊûÅ„ÄÇ
def train_model(texts):
    for comm in tqdm(texts):
        if comm != "":
            text = re.sub(r'(?:ÂõûÂ§ç)?(?://)?@[\w\u2E80-\u9FFF]+:?|\[\w+\]', ',',comm)
            score = SnowNLP(text)
            if score.sentiments > 0.8:
                with open('train_pos.txt', mode='a', encoding='utf-8') as g:
                    g.writelines(comm +"\n")
            elif score.sentiments < 0.3:
                with open('train_neg.txt', mode='a', encoding='utf-8') as f:
                    f.writelines(comm + "\n")
            else:
                pass
```

```python
def test_model(texts):
    for comm in tqdm(texts):
        if comm != "":
            text = re.sub(r'(?:ÂõûÂ§ç)?(?://)?@[\w\u2E80-\u9FFF]+:?|\[\w+\]', ',',comm)
            score = SnowNLP(text)
            if score.sentiments > 0.8:
                with open('test_pos.txt', mode='a', encoding='utf-8') as g:
                    g.writelines(comm +"\n")
            elif score.sentiments < 0.3:
                with open('test_neg.txt', mode='a', encoding='utf-8') as f:
                    f.writelines(comm + "\n")
            else:
                pass
```

```python
# ÂÆö‰πâ‰∏Ä‰∏™ÂáΩÊï∞ËÆ°ÁÆóËØÑËÆ∫ÁöÑÊÉÖÁª™ÂàÜÊï∞
import numpy as np
def snowanalysis(self):
    sentimentslist = []
    for li in self:
        if li != "":   # Êúâ‰∫õÊòØÁ©∫ÁôΩËØÑËÆ∫ÔºåÈúÄË¶ÅË∑≥Ëøá     
            # Ê≠£ÂàôË°®ËææÂºèÊõøÊç¢ÊéâÂõûÂ§çËØÑËÆ∫ÁöÑÂºÄÂ§¥‚ÄúÂõûÂ§ç@‚Ä¶‚Ä¶‚Äù
            li = str(li)  # Áªü‰∏ÄËΩ¨‰∏∫Â≠óÁ¨¶‰∏≤ÈÅøÂÖçÊä•Èîô
            
            if li != 'nan':    # Â≠òÂú®‰∏ÄÈÉ®ÂàÜÁöÑÁ©∫ÁôΩËØÑËÆ∫
                text = re.sub(r'(?:ÂõûÂ§ç)?(?://)?@[\w\u2E80-\u9FFF]+:?|\[\w+\]', '', li)
                # ‰∏≠Ëã±ÊñáÊ†áÁÇπÁ¨¶Âè∑‰ºº‰πéÂ∏¶ÊúâÊÉÖÊÑüÔºåÂÖ®ÈÉ®ÊõøÊç¢Êéâ
                r='[‚ÄôÔºÅÔºüÔºöÔºõ„Äê„ÄëÔºå„Ää„Äã!"#$%&\'()ÔºàÔºâ‚Äú‚Äù‚Ä¶*+,-./:;<=>?@[\\]^_`{|}~]+'
                text = re.sub(r, '', text)
                # ÁªôËØ•Êù°ËØÑËÆ∫ÊâìÂàÜ
                if text:
                    s = SnowNLP(text)
                    # ‰∏∫‰∫ÜÂÅöÈ¢ëÁéáÁõ¥ÊñπÂõæÔºåÊääÂàÜÊï∞‰øùÂ≠òÂú®numpyÊï∞ÁªÑÈáå
                    sentimentslist.append(s.sentiments)
                    score = np.array(sentimentslist)   
    # ËøîÂõûÂàÜÂÄºÁöÑÂàóË°®
    return score
```

```python
import jieba
import re
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB, GaussianNB
from sklearn import metrics
from sklearn.metrics import accuracy_score
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#ËØªÂèñÊñá‰ª∂ÂÜÖÂÆπ
def loadfile(file_dir, label):
    words_list = []
    labels_list = []
    with open(file_dir,'r',encoding='utf-8')as f:
        content = f.readlines()
    for c in content:
        words_list.append(c)
        labels_list.append(label)
    return words_list, labels_list


# Êï∞ÊçÆÈ¢ÑÂ§ÑÁêÜ
def deal_text(text, stop_path):
    stopwords = set()
    with open(stop_path, 'r', encoding='utf-8') as in_file:
        for line in in_file:
            stopwords.add(line.strip())

    text = re.sub('[!ÔºÅ]+', "!", text)
    text = re.sub('[?Ôºü]+', "?", text)
    text = re.sub("[a-zA-Z#$%&\'()*+,-./:;Ôºö<=>@Ôºå„ÄÇ‚òÖ„ÄÅ‚Ä¶„Äê„Äë„Ää„Äã‚Äú‚Äù‚Äò‚Äô[\\]^_`{|}~]+", " UNK ", text)
    text = re.sub(r"\d+", ' NUM ', text)
    text = re.sub(r"\s+", " ", text)
    text = " ".join([term for term in jieba.cut(text) if term and not term in stopwords])
    return text
```

```python
train_words_list1, train_labels1 = loadfile('train_pos.txt', '1')
train_words_list2, train_labels2 = loadfile('train_neg.txt', '0')

train_words_list = train_words_list1 + train_words_list2
train_labels = train_labels1 + train_labels2

test_words_list1, test_labels1 = loadfile('test_pos.txt', '1')
test_words_list2, test_labels2 = loadfile('test_neg.txt', '0')

test_words_list = test_words_list1 + test_words_list2
test_labels = test_labels1 + test_labels2

train_comments_new = [deal_text(comment, "stopwords_cn.txt") for comment in train_words_list]
test_comments_new = [deal_text(comment, "stopwords_cn.txt") for comment in test_words_list]


stop_words = open('stopwords_cn.txt','r',encoding='utf-8').read()
stop_words = stop_words.encode('utf-8').decode('utf-8-sig') #ÂàóË°®Â§¥ÈÉ®\ufeffÂ§ÑÁêÜ
stop_words = stop_words.split('\n') #Ê†πÊçÆÂàÜÈöîÁ¨¶ÂàÜÈöî
```

```python
#ËÆ°ÁÆóÂçïËØçÊùÉÈáç
tf = TfidfVectorizer(stop_words=stop_words,max_df=0.5)

train_features = tf.fit_transform(train_comments_new)
#‰∏äÈù¢fitËøá‰∫Ü,ËøôÈáåtransform
test_features = tf.transform(test_comments_new)

#Â§öÈ°πÂºèË¥ùÂè∂ÊñØÂàÜÁ±ªÂô®
clf = MultinomialNB(alpha=0.001).fit(train_features,train_labels)
predicted_labels = clf.predict(test_features)
#ËÆ°ÁÆóÂáÜÁ°ÆÁéá
print('ÂáÜÁ°ÆÁéáÔºö',accuracy_score(test_labels,predicted_labels))
```
