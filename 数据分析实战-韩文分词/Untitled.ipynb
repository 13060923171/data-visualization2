```python
# coding=utf-8
import jieba
import jieba.analyse
import re
import time
from collections import Counter

#------------------------------------中文分词------------------------------------
cut_words = ""
all_words = ""
f = open('fenci.txt', 'w', encoding='utf-8')
jieba.analyse.set_stop_words('stopwords_cn.txt')   # 停用词词典
for line in sum_content:
    line = str(line)
    line = line.strip('\n')
    #停用词过滤
    line = re.sub('[0-9’!"#$%&\'()*+,-./:;<=>?@，。?★、…【】《》？“”‘’！[\\]^_`{|}~\s]+', "", line)
    seg_list = jieba.cut(line, cut_all=False)
    cut_words = (" ".join(seg_list))

    #计算关键词
    all_words = cut_words.split()
    c = Counter()
    for x in all_words:
        if len(x)>1 and x != '\r\n':
            c[x] += 1
    output = ""
    #print('\n词频统计结果：')
    for (k,v) in c.most_common():
        output += k + " "
    
    f.write(output+"\n")
else:
    f.close()
```

```python
corpus = []

# 读取预料 一行预料为一个文档
for line in open('fenci.txt', 'r',encoding='utf-8').readlines():
    corpus.append(line.strip())
        
# 将文本中的词语转换为词频矩阵 矩阵元素a[i][j] 表示j词在i类文本下的词频
vectorizer = TfidfVectorizer()

# 第一个fit_transform是计算tf-idf 第二个fit_transform是将文本转为词频矩阵
tfidf = vectorizer.fit_transform(corpus)
# 获取词袋模型中的所有词语
word = vectorizer.get_feature_names()

# 将tf-idf矩阵抽取出来 元素w[i][j]表示j词在i类文本中的tf-idf权重
weight = tfidf.toarray()




data = {'word': word,
        'tfidf': weight.sum(axis=0).tolist()}
df = pd.DataFrame(data)
df.head()
# df.sort_values(by="tfidf" , ascending=False) 
# df
```

```python
import string
def str_count(str):
    count_en = count_dg = count_sp = count_zh = count_pu = 0
    for s in str:
        if s in string.ascii_letters:
            count_en += 1
        elif s.isdigit():
            count_dg += 1
        elif s.isspace():
            count_sp += 1
        elif s.isalpha():
            count_zh += 1
        else:
            count_pu += 1
    return count_zh
```

```python
df['zh'] = df['word'].apply(str_count)
```

```python
word1 = []
tfidf1 = []

for w,t,z in zip(df['word'],df['tfidf'],df['zh']):
    if z >= 1:
        word1.append(w)
        tfidf1.append(t)
    else:
        pass
df1 = pd.DataFrame()
df1['word'] = word1
df1['tfidf'] = tfidf1
df1.sort_values(by="tfidf" , ascending=False,inplace=True) 
df1.head()
```

```python
def cut_words():
    # 获取当前文件路径
    counts = {}
    jieba.analyse.set_stop_words('stopwords_cn.txt') 
    for t in sum_content:
        t = str(t)
        if len(t) >=2:
            t = re.sub('[0-9’!"#$%&\'()*+,-./:;<=>?@，。?★、…【】《》？“”‘’！[\\]^_`{|}~\s]+：', "", t)
            text = jieba.lcut(t)
            for word in text:
                if len(word)>1 and word != '\r\n':
                    counts[word] = counts.get(word, 0) + 1
                else:
                    pass
#             for i in range(len(find)):
#                 try:
#                     df2['find'] = [find[i]]
#                     df2['counts'] = [counts[find[i]]]
#                     df2.to_csv('频次.csv',mode='a+',header=None,index=None,encoding='utf-8')
#                 except:
#                     pass
        else:
            pass
    return counts
counts = cut_words()
print(counts)
```

```python
list1 = list(counts.items())
list1.sort(key=lambda x:x[1],reverse=True)
```

```python
def word_count(x):
    for l in list1:
        if x == l[0]:
            return l[1]
```
